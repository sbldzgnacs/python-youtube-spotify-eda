Aykırı Değerler

Analiz
Bir şarkının veya videonun en önemli 3 kriteri izlenme sayısı, aldığı yorum sayısı ve beğeni sayılarıdır.

Bu aykırı değerleri tespit etme işlemini İstatistik bilimindeki IQR (Çeyrekler Arası Açıklık) hesaplaması kullandım

```python
IQR (İnterkuartil Aralık) Yöntemi ile Aykırı değerleri ayıtlıyoruz
```

```python
#Aykırı değerler
Q1 = df['Views'].quantile(0.25) #1.çeyrek
Q3 = df['Views'].quantile(0.75) #3.çeyrek
IQR1 = Q3 - Q1 #çeyrekler arası mesafe

print(IQR1)
```

```python
aykırı_deger = df[(df['Views'] < (Q1 - 1.5 * IQR1)) | (df['Views'] > (Q3 + 1.5 * IQR1))].index
print(aykırı_deger)

#1.5 ile daha geniş aralık bulmak için çarpıyoruz
#Q1 - 1.5 * IQR1   #Alt sınır
# Q3 + 1.5 * IQR1  #Üst Sınır
#outlier_rows1, Views sütunundaki değerlerin alt sınırın altında veya üst sınırın üstünde olduğu satırları seçer
```

```python
df = df[~((df['Views'] < (Q1 - 1.5 * IQR1)) | (df['Views'] > (Q3 + 1.5 * IQR1)))]


#aykırı değerlerin tersini (alt ve üst sınırların içinde kalan değerleri) seçer.
#aykırı değerleri çıkartıyoruz
```

```python
# group by artist and sum the views and streams
df_grouped = df.groupby('Artist')[['Views', 'Stream']].mean().reset_index()

# sort the data by views in descending order
df_grouped = df_grouped.sort_values(by=[
                           "Stream", "Views"], ascending=[False, False])[:10]

# melt the data to create a long format
df_melted = pd.melt(df_grouped, id_vars=['Artist'], var_name='Platform', value_name='Counts')

# create a grouped barplot
sns.barplot(y='Artist', x='Counts', hue='Platform', data=df_melted, orient = 'h')
plt.xticks(rotation=0)
plt.xlabel('Artist')
plt.ylabel('Counts')
plt.title('Sanatçılara Göre Toplam Spotify ve Youtube Dinlenme Sayıları Top 10 ')
plt.show()
```

```python
#Stream(Spotify) - Views(Youtube) sanatçılara göre dinlenme sayılarını görüyoruz
```

YouTube, hem müzik hem de video içeriği sunan bir platformdur. Bu, kullanıcıların müzik dinlemek yerine video izlemeye yönelmesine neden olabilir.
